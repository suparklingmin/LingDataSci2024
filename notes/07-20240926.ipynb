```python
import numpy as np
scores = np.array(
    [[[80, 75, 90],
      [84, 81, 93],
      [89, 72, 98],
      [88, 85, 97]],

     [[91, 83, 79],
      [84, 99, 82],
      [92, 90, 85],
      [93, 92, 80]]]
)
```

```python
%pip install numpy==1.26.4
```

```python
axis0 = ['강은수', '조재영'] # 학생
axis1 = ['3월', '4월', '5월', '6월'] # 월
axis2 = ['국어', '영어', '수학'] # 과목

exam = []
for (i, student) in enumerate(axis0):
    for (j, month) in enumerate(axis1):
        for (k, subject) in enumerate(axis2):
            exam.append((student, month, subject, scores[i,j,k]))

exam
```

```python
%pip install pandas openpyxl
```

```python
import pandas as pd
df_exam = pd.DataFrame(exam, columns=['학생', '월', '과목', '점수'])
df_exam
```

```python
df_exam[(df_exam['점수'] > 85) & (df_exam['과목'] == '영어')]
```

```python
df_exam['점수'].mean()
```

```python
df_exam.groupby(['학생'])['점수'].mean()
```

```python
df_exam.groupby(['학생', '월'])['점수'].mean()
```
